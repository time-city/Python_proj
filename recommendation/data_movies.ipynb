```python
import pandas as pd 

df = pd.read_csv("movies_metadata.csv")
df.head()
```

```python
df.columns
```

```python
columns_nonused = ["adult", "belongs_to_collection", "budget", "homepage", "original_title", "popularity", "production_companies", "production_countries", "revenue", "spoken_languages", "status", "vote_count", "video", "imdb_id"]
movies = df.drop(columns= columns_nonused)
movies.head()
```

```python
columns_rename = {"overview": "description", "vote_average": "rating", "runtime": "duration"}
movies = movies.rename(columns= columns_rename)
movies.head()
```

```python
import ast

def extract_genres(genres_str):
    try:
        genres_list = ast.literal_eval(genres_str)  # parse string → list of dict
        return ', '.join([g['name'] for g in genres_list])
    except:
        return ''

movies['genres'] = movies['genres'].apply(extract_genres)

# Kiểm tra kết quả
movies['genres'].head()
```

```python
movies.head()
```

```python
movies.isna().sum()
```

```python

cols_to_check = ['genres', 'original_language', 'description', 
                 'poster_path', 'release_date', 'duration', 
                 'title', 'rating']

movies_clean = movies.dropna(subset=cols_to_check)

print(f"Before: {len(movies)} rows")
print(f"After:   {len(movies_clean)} rows")
print(f"Deleted: {len(movies) - len(movies_clean)} rows")
```

```python
movies_clean.isna().sum()
```

```python
movies_clean = movies_clean.drop(columns="tagline")
movies_clean.head()
```

```python
movies_clean.to_csv("movies_final.csv", index = False)
```
